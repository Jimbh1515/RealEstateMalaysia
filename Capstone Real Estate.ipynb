```python
## Cleaning the Price and Size Columns
df = pd.read_csv("kl_properties.csv")
df["Price"] = pd.to_numeric(df['Price'], errors='coerce')
df["Size"] = pd.to_numeric(df['Size'], errors='coerce')
print(df.head(10))
```

```python
## Determining any outliers##
z_scores_y = np.abs((df["Price (MYR)"] - df["Price (MYR)"].mean()) / df["Price (MYR)"].std())
z_scores_x = np.abs((df["Size (sqft)"] - df["Size (sqft)"].mean()) / df["Size (sqft)"].std())
threshold = 3

##Filter any outliers##
filtered_x = ["Size (sqft)"[i] for i in range(len("Size (sqft)")) if z_scores_y[i] < threshold and z_scores_y[i] < threshold]
filtered_y = ["Price (MYR)"[i] for i in range(len("Price (MYR)")) if z_scores_y[i] < threshold and z_scores_y[i] < threshold]

df_filtered = df[(z_scores_y < threshold) & (z_scores_x < threshold)]
print(df_filtered.head(10))
print(df_filtered.shape)
```

```python
sns.scatterplot(x= "Size (sqft)", y = "Price (MYR)", data = df_filtered, hue = "Area")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (MYR)")
plt.title("Size vs Price")
plt.show()
```

```python
sns.boxplot(x = "Area", y = "Price (MYR)", data = df_filtered, showmeans=True, showfliers=False)
plt.xlabel("Area")
plt.ylabel("Price (MYR)")
plt.title("Area vs Price")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("kl_properties_cleaned.csv")
## Determining any outliers##
z_scores_y = np.abs((df["Price (MYR)"] - df["Price (MYR)"].mean()) / df["Price (MYR)"].std())
z_scores_x = np.abs((df["Size (sqft)"] - df["Size (sqft)"].mean()) / df["Size (sqft)"].std())
threshold = 3

##Filter any outliers##
filtered_x = ["Size (sqft)"[i] for i in range(len("Size (sqft)")) if z_scores_y[i] < threshold and z_scores_y[i] < threshold]
filtered_y = ["Price (MYR)"[i] for i in range(len("Price (MYR)")) if z_scores_y[i] < threshold and z_scores_y[i] < threshold]


df_filtered = df[(z_scores_y < threshold) & (z_scores_x < threshold)]

sns.relplot(x = "Size (sqft)", y = "Price (MYR)", data = df_filtered, kind = "scatter", style = "Area", hue = "Area")
plt.xlabel("Size (sqft)")
plt.ylabel("Price (MYR)")
plt.title("Size vs Price")
```

```python
sns.relplot(x = "Bedrooms", y = "Price (MYR)", data = df_filtered, kind = "scatter", style = "Area", hue = "Area")
print(df_filtered["Price (MYR)"].corr(df_filtered["Bedrooms"]))
plt.show()
```

```python
sns.relplot(x = "Size (sqft)", y = "Price (MYR)", data = df_filtered, kind = "scatter", col = "Bedrooms", hue = "Area")
print(df_filtered["Price (MYR)"].corr(df_filtered["Size (sqft)"]))
print(df_filtered["Price (MYR)"].corr(df_filtered["Bathrooms"]))
plt.show()
```
